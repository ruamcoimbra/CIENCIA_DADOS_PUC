## Aprendendo Pandas na Prática

[Stack Overflow Annual Developer Survey](https://insights.stackoverflow.com/survey) é o maior e mais abrangente pesquisa sobre desevenvolvedores ao redor do mundo. Considerando o conjunto de dados da pesquisa referente ao ano de 2018, faremos algumas tarefas rotineiras de um cientista de dados utilizando a biblioteca Pandas. É importante lembrar que os resultados da pesqusia foram anonimizados e distribuídos para download publicamente sob a licença [Open Database Licence (ODbL)](https://opendatacommons.org/licenses/odbl/1.0/)

```python
# Dataset: tack Overflow Annual Developer Survey - 2018
import numpy as np
import pandas as pd
print("Importa Numpy e Pandas \n")
```

```python
# 1. Carregue o dataset para análise, disponível na pasta data, usando as funções de leitura do Pandas.
print("Execicio 1 \n")
print("Carrega CSV \n")
data = pd.read_csv("Data\survey_results_public.csv", dtype = object)
```

```python
# 2. Obtenha informações do Data Frame carregado em memória
print("Execicio 2 \n")
print(data.info())
print("\n")
```

```python
# 3. Selecione todas as linhas com país igual a Brasil
print("Execicio 3 \n")
dadosBrasil = data.loc[data['Country'] == "Brazil"]
print(dadosBrasil)
print("\n")
```

```python
# 3.1. Obtenha informações do dadosBrasil carregado em memória
print("Execicio 3.1 \n")
print(dadosBrasil.info())
print("\n")
```

```python
# 4. Os dados do brasil possui dados ausentes (missing data)? Se sim, selecione todas as linhas
#    que contenham algum dado ausente
#    any(axis=1) = Qualquer coluna que tenha valor ausente ou NaN
#    any(axis=0) = Qualquer linha  que tenha valor ausente ou NaN
print("Execicio 4 \n")
dadosBrasil_Null = dadosBrasil[dadosBrasil.isnull().any(axis=1)]
print(dadosBrasil_Null)
print("\n")
```

```python
# 4.1. Obtenha informações do dadosBrasil carregado em memória
print("Execicio 4.1 \n")
print(dadosBrasil_Null.info())
print("\n")
```

```python
# 5. Qual o total de valores ausentes para cada coluna?
print("Execicio 5 \n")
print(dadosBrasil.isnull().sum())
print("\n")
```

```python
# 6. Remova linhas com pelo menos 1 elemento ausente das colunas: 'FormalEducation', 'UndergradMajor', 'Age'.
print("Execicio 6 \n")
limpadadosBrasil = dadosBrasil.dropna(subset=["FormalEducation","UndergradMajor","Age"])
print(limpadadosBrasil)
print("\n")
```

```python
# 6.1. Obtenha informações do dadosBrasil carregado em memória
print("Execicio 6.1 \n")
print(limpadadosBrasil.info())
print("\n")
```

```python
# 7. Crie um subconjunto de dados com as colunas: 
# Respondent, Country, Student, Employment, FormalEducation, UndergradMajor, Exercise, Age
print("Execicio 7 \n")
analiseDados = limpadadosBrasil.filter(["Respondent", "Country", "Student", "Employment", "FormalEducation", "UndergradMajor", "Exercise", "Age"])
print(analiseDados)
print("\n")
```

```python
# 7.1. Obtenha informações do dadosBrasil carregado em memória
print("Execicio 7.1 \n")
print(analiseDados.info())
print("\n")
```
